```python
car_name = []
model_name = []
kilometre = []
price_list = []
features = []
for i in range(1,35):
    #URL="https://www.carsome.my/buy-car".format(i)
    URL="https://www.carsome.my/buy-car?pageNo="+str(i)
    CARSOME = requests.get(URL)
    pagecontent = CARSOME.text
    soup = BeautifulSoup(pagecontent)
    list_1 = soup.find_all('div',attrs={"class":"mod-card__content"})

    for j in list_1:
        name = j.find('a',attrs={"class":"mod-card__title"})
        if name is None:
            car_name.append(np.NaN)
        else:
            car_name.append(name.text)
            
        model = j.find('div',attrs={'class':"mod-card__title__variant-engine"})
        if model is None:
            model_name.append(np.NaN)
        else:
            model_name.append(model.text) 
        km = j.find('div',attrs={'class':"mod-card__car-other"})
        if km is None:
            kilometre.append(np.NaN)
        else:
            kilometre.append(km.text)
            
        price = j.find('div',attrs={'class':"mod-card__price"})
        if price is None:
            price_list.append(np.NaN)
        else:
            price_list.append(price.text)
            
        feature = j.find('div',attrs={'class':"mod-card__high-tag-list"})
        if price is None:
            features.append(np.NaN)
        else:
            features.append(feature.text) 
```

```python
carsome = pd.DataFrame({'carsname':car_name,'modelname':model_name,'kilometres':kilometre,'price':price_list,'features':features})
```
