```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

```python
review = pd.read_csv(r"D:/saikrishna data science/MACHINE TEXT PROJECT/amazon_food_reviews/amazon_food_reviews.csv")
```

```python
review
```

```python
review.rename(columns={"Unnamed: 0":'Id','0':'ProductId','1':'UserId','2':'ProfileName','3':'HelpfulnessNumerator','4':'HelpfulnessDenominator','5':'Score','6':'Time','7':'ReviewSummary','8':'ReviewText'},inplace=True)
```

```python
review.head()
```

```python
review.set_index('Id', inplace=True)
```

```python
review.head()
```

```python
UserId = []
for i in review['UserId']:
    UserId.append(i.split()[1])
```

```python
review['UserId'] = UserId
```

```python
review.head()
```

```python
ProductId = []
for i in review['ProductId']:
    ProductId.append(i.split()[1])
    
review['ProductId'] = ProductId    
```

```python
review.head()
```

```python
HelpfulnessNumerator = []
for i in review['HelpfulnessNumerator']:
    HelpfulnessNumerator.append(i.split()[1])
    
review['HelpfulnessNumerator'] = HelpfulnessNumerator    
```

```python
review.head()
```

```python
HelpfulnessDenominator = []
for i in review['HelpfulnessDenominator']:
    HelpfulnessDenominator.append(i.split()[1])
    
review['HelpfulnessDenominator'] = HelpfulnessDenominator
```

```python
Score = []
for i in review['Score']:
    Score.append(i.split()[1])
    
review['Score'] = Score
```

```python
Time = []
for i in review['Time']:
    Time.append(i.split()[1])
    
review['Time'] = Time
```

```python
review.head()
```

```python
review['ReviewText'] = review['ReviewText'].apply(lambda x:x.replace('ReviewText:',""))
```

```python
review.head()
```

```python
review['ProfileName'] = review['ProfileName'].apply(lambda x:x.replace('ProfileName:',""))
review['ProfileName'] = review['ProfileName'].apply(lambda x:x.replace('\n',""))
```

```python
review.head()
```

```python
review['ReviewSummary'] = review['ReviewSummary'].apply(lambda x:x.replace('ReviewSummary:',""))
review['ReviewSummary'] = review['ReviewSummary'].apply(lambda x:x.replace('\n',""))
```

```python
review.head()
```

```python
review.to_csv("review.csv",index=False)
```

```python
review.shape
```

```python
review.info()
```

```python
review['HelpfulnessNumerator'] = review['HelpfulnessNumerator'].astype(str).astype('int64')
```

```python
review['HelpfulnessDenominator'] = review['HelpfulnessDenominator'].astype(str).astype('int64')
```

```python
review['Score'] = review['Score'].astype(str).astype('int64')
```

```python
amzon_review = review.drop_duplicates()
```

```python
amzon_review.info()
```

```python
sentiment = []
for i in amzon_review['Score']:
    if i > 3:
        sentiment.append('positive')
    elif i < 3:
        sentiment.append('negative')
    else:
        sentiment.append('not')
amzon_review['sentiment'] = sentiment        
```

```python
amzon_review.head()
```

```python
amzon_review['Helpfulness'] = amzon_review['HelpfulnessNumerator']/amzon_review['HelpfulnessDenominator']
```

```python
usefulness = []
for i in amzon_review['Helpfulness']:
    if i > 0.75:
        usefulness.append(">75%")
    elif i < 0.25:
        usefulness.append("<25%")
    elif i >= 0.25 and i <= 0.75:
        usefulness.append("25-75%")
    else:
        usefulness.append("useless")
amzon_review['Helpfulness']  = usefulness      
```

```python
word_count = []
for i in amzon_review['ReviewText']:
    word_count.append(len(i.split()))
amzon_review['word_count'] = word_count    
```

```python
amzon_review.head()
```

```python
pos = amzon_review[amzon_review['sentiment']=='positive']
```

```python
pos
```

```python
neg = amzon_review[amzon_review['sentiment']=='negative']
neg
```

### Popular words in Positive Reviews (4-5 Rating)


```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
comment_words = ''
stopwords = set(STOPWORDS)


for val in pos['ReviewSummary']:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    comment_words += " ".join(tokens)+" "
     
    
wordcloud = WordCloud(width = 600, height = 600,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)    
    
plt.figure(figsize = (10, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()  
```

### Popular words in Negative Reviews (1-2 Rating)

```python
comment_words = ''
stopwords = set(STOPWORDS)

for val in neg['ReviewSummary']:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    comment_words += " ".join(tokens)+" "
    
wordcloud = WordCloud(width = 600, height = 600,
                background_color ='black',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)    
    
plt.figure(figsize = (10, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
