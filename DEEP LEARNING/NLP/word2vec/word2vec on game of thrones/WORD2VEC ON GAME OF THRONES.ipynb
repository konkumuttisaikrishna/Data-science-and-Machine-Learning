```python
import gensim
import os
```

```python
# import natural language toolkit
import nltk
# download the punkt tokenizer
nltk.download('punkt')
print("The punkt tokenizer is downloaded")
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('Untitled Folder'):
    
    f = open(os.path.join('Untitled Folder',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
# Import necessary libraries
from gensim.models import Word2Vec
# Create a Word2Vec model
model = Word2Vec(story, vector_size=300, window=7, min_count=3, sg=1)
```

```python
# Training the Word2Vec model
model.train(story, total_examples=len(story), epochs=10)
```

```python
# find top N similar word to a given word in the vocabulary
print(model.wv.most_similar("targaryen",topn=5))
```
