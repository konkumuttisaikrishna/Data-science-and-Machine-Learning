### REMOVING OUTLIERS FROM THE DATA.

```python
def out_rem(column):
    q3 = concrete[column].quantile(q=0.75)
    q1 = concrete[column].quantile(q=0.25)
    iqr = q3 - q1
    upper_bound = q3 + 1.5*iqr
    lower_bound = q1 - 1.5*iqr
    out_idx = concrete[(concrete[column]>upper_bound) | (concrete[column]<lower_bound)].index
    print(concrete[column].name)
    print(q3,"-",q1,"=",iqr)
    print(upper_bound)
    print(lower_bound)
    print(out_idx)
    concrete.drop(labels=(out_idx),axis=0,inplace=True)
```

### Variance Infllation Factor (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif
```

```python
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
np.round(lr.pvalues, 3)
```
